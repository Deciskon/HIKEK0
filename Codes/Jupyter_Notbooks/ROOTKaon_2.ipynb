```c++
//PARAMETER ESTIMATION: CS, Ccos, Csin

void meat_CcCs(int bins, int noh, int endd, int events, double DiluCon, double Cint, double Phi0)
{
    double CL = 1.0, CS = 0.43;
    double tauKS = 0.08954, tauKL = 51.16;
    double MK = 497.614 /*in MeV/c^2*/,  dMK = 2.1890e-11 /*in MeV/c^2*/;
    //double Cint = 0.12,  Phi0 = 0.201357921 /*Phi0 = arccos(sqrt(0.96))*/;
    //double DiluCon = 0.285714286 /*Dilucon = 2.0/7.0;   N(K0)/N(K0bar) = 1.8*/;
    double Csin = (Cint*DiluCon*sin(Phi0)), Ccos = (Cint*DiluCon*cos(Phi0)); //C_f(x) = Cint*Dilucon*f(Phi0)
    //Csin = 0.00685714   ;   Ccos = 0.033593
    
    double mom = 75000.0; //in MeV/c
    double c_light = 299.792; //in 10^6 m/s
    double tauK = 2.0/((1.0/tauKS)+(1.0/tauKL)), gamma = mom/MK;

    double lambdaL = gamma*c_light*tauKL, lambdaS = gamma*c_light*tauKS, lambda = gamma*c_light*tauK;

    ostringstream decayform;
    decayform<<CL<<"*exp(-x*"<<tauKS/tauKL<<") + "<<CS<<"*exp(-x*"<<tauKS/tauKS<<") + 2*exp(-x*"<<tauKS/tauK<<")*"<<DiluCon<<"*"<<Cint<<"*cos("<<Phi0<<"-("<<dMK<<"*x/"<<tauKS<<"e-9))";                            
    //decayform<<CL<<"*exp(-x*"<<tauKS/tauKL<<")";
    //decayform<<CS<<"*exp(-x*"<<tauKS/tauKS<<")";
    //decayform<<CL<<"*exp(-x*"<<tauKS/tauKL<<") + "<<CS<<"*exp(-x*"<<tauKS/tauKS<<")";
    //decayform<<"2*exp(-x*"<<tauKS/tauK<<")*"<<DiluCon<<"*"<<Cint<<"*cos("<<Phi0<<"-("<<dMK<<"*x/"<<tauKS<<"e-9))";                            
    
    int i=0, j=0;
    TF1 *decayf = new TF1("decayf",decayform.str().c_str(),0,endd);

    ostringstream decayfit;
    decayfit<<"[3] * ("<<CL<<"*exp(-x*"<<tauKS/tauKL<<") + [0]*exp(-x*"<<tauKS/tauKS<<") + 2*exp(-x*"<<tauKS/tauK<<")*(([1]*cos("<<dMK<<"*x/"<<tauKS<<"e-9)) + ([2]*sin("<<dMK<<"*x/"<<tauKS<<"e-9))))";                      
    
    //decayfit<<"[3] * ("<<CL<<"*exp(-x*"<<tauKS/tauKL<<") + [0]*exp(-x*"<<tauKS/tauKS<<") + 2*exp(-x*"<<tauKS/tauK<<")*(("<<Ccos<<"*cos("<<dMK<<"*x/"<<tauKS<<"e-9)) + ("<<Csin<<"*sin("<<dMK<<"*x/"<<tauKS<<"e-9))))";                      
    //decayfit<<"[3] * ("<<CL<<"*exp(-x*"<<tauKS/tauKL<<") + "<<CS<<"*exp(-x*"<<tauKS/tauKS<<") + 2*exp(-x*"<<tauKS/tauK<<")*(([1]*cos("<<dMK<<"*x/"<<tauKS<<"e-9)) + ("<<Csin<<"*sin("<<dMK<<"*x/"<<tauKS<<"e-9))))";                      
    //decayfit<<"[3] * ("<<CL<<"*exp(-x*"<<tauKS/tauKL<<") + "<<CS<<"*exp(-x*"<<tauKS/tauKS<<") + 2*exp(-x*"<<tauKS/tauK<<")*(("<<Ccos<<"*cos("<<dMK<<"*x/"<<tauKS<<"e-9)) + ([2]*sin("<<dMK<<"*x/"<<tauKS<<"e-9))))";                      
    
    //decayfit<<"[3] * ("<<CL<<"*exp(-x*"<<tauKS/tauKL<<") + [0]*exp(-x*"<<tauKS/tauKS<<") + 2*exp(-x*"<<tauKS/tauK<<")*(([1]*cos("<<dMK<<"*x/"<<tauKS<<"e-9)) + ("<<Csin<<"*sin("<<dMK<<"*x/"<<tauKS<<"e-9))))";                      
    //decayfit<<"[3] * ("<<CL<<"*exp(-x*"<<tauKS/tauKL<<") + [0]*exp(-x*"<<tauKS/tauKS<<") + 2*exp(-x*"<<tauKS/tauK<<")*(("<<Ccos<<"*cos("<<dMK<<"*x/"<<tauKS<<"e-9)) + ([2]*sin("<<dMK<<"*x/"<<tauKS<<"e-9))))";                      
    
    //decayfit<<"[3] * ("<<CL<<"*exp(-x*"<<tauKS/tauKL<<") + [0]*exp(-x*"<<tauKS/tauKS<<")";
    
    TF1 fot("fot", decayfit.str().c_str());
    fot.SetParameter(0,0.43);
    fot.SetParameter(1,0.033593);
    fot.SetParameter(2,0.00685714);
    TH1D CShist("CShist","Parameter Estimation: C_{S};C_{S};Count",100,-2.5,2.5);
    TH1D Ccoshist("Ccoshist","Parameter Estimation: C_{cos};C_{cos};Count",100,-2.5,2.5);
    TH1D Csinhist("Csinhist","Parameter Estimation: C_{sin};C_{sin};Count",100,-2.5,2.5);
    double CSmean[noh], CSsd[noh];           //[0]
    double Ccosmean[noh], Ccossd[noh];      //[1]
    double Csinmean[noh], Csinsd[noh];     //[2]
    double xtau[bins], extau[bins], ytau[bins], eytau[bins], p3 = 0, p3mean[noh];

    for(j=0;j<bins;j++)
    {
        extau[j] = 0;
        xtau[j] = (double(endd)/double(bins)) * (double(j) + 0.5);
        eytau[j] = 0;
        ytau[j] = 0;
    }
    
    for(i=0;i<noh;i++)
    {
        TH1D *h=new TH1D("h","Random K_{0} Decay Histogram;t [#tau_{S}];Count",bins,0,endd);
        for(j=0;j<events;j++)
        {
            h->Fill(decayf->GetRandom());
        }
        h->SetMinimum(0.0);
        h->Fit("fot","q");
        TF1 *glove = h->GetFunction("fot");
        p3 = glove->GetParameter(3);

        CSmean[i]=glove->GetParameter(0);
        CSsd[i]=glove->GetParError(0);
        CShist.Fill(CSmean[i]);

        Ccosmean[i]=glove->GetParameter(1);
        Ccossd[i]=glove->GetParError(1);
        Ccoshist.Fill(Ccosmean[i]);    

        Csinmean[i]=glove->GetParameter(2);
        Csinsd[i]=glove->GetParError(2);
        Csinhist.Fill(Csinmean[i]);

        
        for(j=0;j<bins;j++)
        {
            double *xar = new double(xtau[j]);
            ytau[j] += glove->EvalPar(xar) / p3;
            eytau[j] += (glove->EvalPar(xar)/p3) * (glove->EvalPar(xar)/p3);
            //eytau[j] += glove->GetParameter(3) *  ((exp(-(*xar))*CSsd[i]) + (2*DiluCon*exp(-(*xar)*0.500875)*((cos(0.201357921)*cos(2.189e-11*(*xar)*0.08954)) + (sin(0.201357921)*sin(2.189e-11*(*xar)*0.08954)))*Cintsd[i]));
        }
      
        h->Delete();
    }
    
    gStyle->SetOptStat(1111111);

    CShist.Draw();
    gPad->Print("CS__CcCs.png");

    Ccoshist.Draw();
    gPad->Print("Ccos.png");

    Csinhist.Draw();
    gPad->Print("Csin.png");
    
    
    for(j=0;j<bins;j++)
    {
        ytau[j] = (ytau[j]/double(noh)) ;
        eytau[j] = sqrt((eytau[j]/double(noh)) - (ytau[j]*ytau[j]));//(eytau[j]/double(noh));
    }
    TGraphErrors gr(bins,xtau,ytau,extau,eytau);
    ostringstream titl;
    titl<<"Fitted Decay Function:   n(Fits)="<<noh<<", n(Bins)="<<bins<<", n(Events)="<<events<<";t [#tau_{S}];f(t)";
    gr.SetTitle(titl.str().c_str());
    //gr.SetMinimum(0);
    //gr.SetMaximum(1300);
    //gr.SetFillStyle(3002);
    //gr.SetFillColor(66);
    gr.SetMarkerColor(kBlack);
    gr.SetMarkerSize(0.4);
    gr.SetMarkerStyle(8);
    gr.SetLineColor(kRed);
    gr.SetMaximum(2.0);
    gr.SetMinimum(0.5);
    gr.Draw("APC");
    gPad->Print("Fit_w_Errors__CcCs.png");
    
    cout<<"\nCS = "<<CS<<"\nCcos = "<<Ccos<<"\nCsin = "<<Csin<<"\n";
}




// //V O I D    M A I N
// int bins = 60, noh = 400, endd = 6, events = (1830*8);
// double DiluCon = 0.285714286;//0.285714286;   //Dilucon = 2.0/7.0;  N(K0)/N(K0bar) = 1.8
// double Cint = 0.12;//0.12;
// double Phi0 = 0.201357921;//0.201357921;
// meat_CcCs(bins, noh, endd, events, DiluCon, Cint, Phi0);
```

```c++
//ROOFit Parameter Estimation - {Cs,Cint,Phi0} with Binned Data

double roof_CiPh(int bins, int noh, double startt, double endd, int events, double DiluCon, double CSv, double Cintv, double Phi0v, double CLfrac, int sorint)
{
    using namespace RooFit;
    
    double CL = 1.0;
    double tauKS = 0.08954, tauKL = 51.16; //in nano-seconds
    double MK = 497.614 /*in MeV/c^2*/,  dMK = 2.1890e-11 /*in MeV/c^2*/;
    //double Cint = 0.12,  Phi0 = 0.201357921 /*Phi0 = arccos(sqrt(0.96))*/;
    //double DiluCon = 0.5;//0.285714286 /*Dilucon = 2.0/7.0;   N(K0)/N(K0bar) = 1.8*/;

    //double CS = 0.43, Csin = (Cint*DiluCon*sin(Phi0)), Ccos = (Cint*DiluCon*cos(Phi0)); //C_f(x) = Cint*Dilucon*f(Phi0)
    //Csin = 0.00685714   ;   Ccos = 0.033593

    double mom = 75000.0; //in MeV/c
    double c_light = 299.792; //in 10^6 m/s
    double tauK = 2.0/((1.0/tauKS)+(1.0/tauKL)), gamma = mom/MK;

    double lambdaL = gamma*c_light*tauKL, lambdaS = gamma*c_light*tauKS, lambda = gamma*c_light*tauK;

    ostringstream decayform, decayfit;
    decayform<<CL<<"*exp(-x*"<<tauKS/tauKL<<") + CS*exp(-x*"<<tauKS/tauKS<<") + 2*exp(-x*"<<tauKS/tauK<<")*"<<DiluCon<<"*Cint*cos(Phi0-("<<dMK<<"*x/"<<tauKS<<"e-9))";                            
    //decayform<<CL<<"*exp(-x*"<<tauKS/tauKL<<")";
    //decayform<<"CS*exp(-x*"<<tauKS/tauKS<<")";
    //decayform<<CL<<"*exp(-x*"<<tauKS/tauKL<<") + CS*exp(-x*"<<tauKS/tauKS<<")";
    //decayform<<"2*exp(-x*"<<tauKS/tauK<<")*"<<DiluCon<<"*Cint*cos(Phi0-("<<dMK<<"*x/"<<tauKS<<"e-9))";
    //decayform<<CL<<"*exp(-x*"<<tauKS/tauKL<<") + 2*exp(-x*"<<tauKS/tauK<<")*"<<DiluCon<<"*Cint*cos(Phi0-("<<dMK<<"*x/"<<tauKS<<"e-9))";
    //decayform<<"CS*exp(-x*"<<tauKS/tauKS<<") + 2*exp(-x*"<<tauKS/tauK<<")*"<<DiluCon<<"*Cint*cos(Phi0-("<<dMK<<"*x/"<<tauKS<<"e-9))";
    //decayform<<"0.1 + CS*exp(-x*"<<tauKS/tauKS<<") + 2*exp(-x*"<<tauKS/tauK<<")*"<<DiluCon<<"*Cint*cos(Phi0-("<<dMK<<"*x/"<<tauKS<<"e-9))";                            
    //decayform<<"1";
    
    
    //////////////////////////////////////////CLendd Evaluation////////////////////////////////////////////////////////
    ostringstream cldecayform;
    cldecayform<<CL<<"*exp(-x*"<<tauKS/tauKL<<")";
    TF1 *cldecayf = new TF1("cldecayf",cldecayform.str().c_str(),startt,endd);
    double CLendd = cldecayf->EvalPar(&endd);
    /////////////////////////////////////////////////////////////////////////////////////////////////////////////////////
    
    
    
    int i=0, j=0;

    TH1D CShist("CShist","Parameter Estimation: C_{S};C_{S};Count",100,-0.5,1.5);
    TH1D Cinthist("Cinthist","Parameter Estimation: C_{int};C_{int};Count",100,-0.5,1.5);
    TH1D Phi0hist("Phi0hist","Parameter Estimation: #phi_{0};#phi_{0};Count",100,0.0,0.4);//-1.6,1.6);
    
    RooRealVar x("x", "x", startt, endd);
    RooRealVar CS("CS", "CS", CSv, 0.00, 1.0);
    if(sorint==0)
    {
        CS.setConstant(true);
    }
    RooRealVar Cint("Cint", "Cint", Cintv, 0.00, 1.0);
    if(sorint==1)
    {
       Cint.setConstant(true); 
    }
    RooRealVar Phi0("Phi0", "Phi0", Phi0v, -1.58, 1.58);
    //Phi0.setConstant(true);
    RooGenericPdf decaypdf("decaypdf", "decaypdf", decayform.str().c_str(), RooArgSet(x, CS, Cint, Phi0));
    
    RooGaussian Phicon("Phicon","Phicon", Phi0, RooConst(Phi0v), RooConst(0.1/*From Precision Measurement Paper*/)) ;

    //decayfit<<"CS*exp(-x*"<<tauKS/tauKS<<") + 2*exp(-x*"<<tauKS/tauK<<")*"<<DiluCon<<"*Cint*cos(Phi0-("<<dMK<<"*x/"<<tauKS<<"e-9))";
    decayfit<<CL*CLfrac<<"*exp(-x*"<<tauKS/tauKL<<") + CS*exp(-x*"<<tauKS/tauKS<<") + 2*exp(-x*"<<tauKS/tauK<<")*"<<DiluCon<<"*Cint*cos(Phi0-("<<dMK<<"*x/"<<tauKS<<"e-9))";                            
    //decayfit<<"CS*Cint*Phi0";                            
    RooGenericPdf woaht("woaht", "woaht", decayfit.str().c_str(), RooArgSet(x, CS, Cint, Phi0));
    
    
    
    double corravg = 0;
    for(i=0;i<noh;i++)
    {
        CS.setVal(CSv);
        Cint.setVal(Cintv);
        Phi0.setVal(gRandom->Gaus(Phi0v,0.1));
        
        RooDataSet *data = decaypdf.generate(x, events);
        
        
        TH1 *roohist = data->createHistogram("roohist", x, Binning(bins,startt,endd));
        for(j=1;j<=bins;j++)
        {
            roohist->SetBinContent(j, (roohist->GetBinContent(j)-((CL-CLfrac)*roohist->GetBinContent(bins)*(CL/CLendd)*exp(-(j*(endd/bins))*tauKS/tauKL))));
            if(roohist->GetBinContent(j) < 0.0)
            {
                roohist->SetBinContent(j, 0.0);
            }
        }
        RooDataHist rdh("rdh", "rdh", x, roohist);
        
        //std::unique_ptr<RooFitResult> fitResult{decaypdf.fitTo(rdh, ExternalConstraints(Phicon), Save(), PrintLevel(-1))};
        std::unique_ptr<RooFitResult> fitResult{woaht.fitTo(rdh, ExternalConstraints(Phicon), Save(), PrintLevel(-1))};
        //fitResult->Print();
      
        CShist.Fill(CS.getValV());
        Cinthist.Fill(Cint.getValV());
        Phi0hist.Fill(abs(Phi0.getValV()));
        
        corravg += (fitResult->correlation("CS","Cint")/double(noh));
       
        if(i==(noh-1))
        {
            RooPlot *xframe = x.frame(Title("K0 Decay PDF"));
            rdh.plotOn(xframe);
            woaht.plotOn(xframe);
            xframe->Draw();
            gPad->Print("ROOHIST.png");
        }   
    }
    
    cout<<"\n\n";
    gStyle->SetOptStat(1111111);
    CShist.Draw();
    gPad->Print("Roofit_CS.png");
    Cinthist.Draw();
    gPad->Print("Roofit_Cint.png");
    Phi0hist.Draw();
    gPad->Print("Roofit_Phi0.png");
    cout<<"CS = "<<CSv<<"\nCint = "<<Cintv<<"\nPhi0 = "<<Phi0v<<"\n\nAvgCorr{CS,Cint} = "<<corravg<<"\n\n";
    
    
    if(sorint==0)
    {
        return (Cinthist.GetMean());
    }
    if(sorint==1)
    {
        return (CShist.GetMean());
    }
    return 0;
}
```

```c++
//V O I D    M A I N
int bins = 60, noh = 100, events = (50000);
double startt = 0, endd = 12;
double DiluCon = 0.285714286;//0.285714286;   //Dilucon = 2.0/7.0;  N(K0)/N(K0bar) = 1.8
double CS = 0.43;//+-0.008455;//0.43
double Cint = 0.12;//+-0.01592;//0.12;
double Phi0 = 0.201357921;//0.201357921;

double CLfrac = 1.0;
int sorint = 0; //0 to fix CS, 1 to fix Cint, otherwie to fix neither

int i, j, loops = 15;
for(i=0;i<loops;i++)
{
    sorint = (i%2);
    if(sorint==0)
    {
        endd = 12;
        Cint = roof_CiPh(bins, noh, startt, endd, events, DiluCon, CS, Cint, Phi0, CLfrac, sorint); 
    }
    if(sorint==1)
    {
        endd = 6;
        CS = roof_CiPh(bins, noh, startt, endd, events, DiluCon, CS, Cint, Phi0, CLfrac, sorint); 
    }
}
```

```c++
//ROOFit Parameter Estimation - {Cs,Cint,Phi0} with Recursive Optimization

std::array<double,2> roof_CiPh2(int bins, double startt, double endd, int events, double DiluCon, double CSv, double Cintv, double Phi0v, double CLfrac, int loops)
{
    using namespace RooFit;
    int sorint = 0; //0 to fix CS, 1 to fix Cint, otherwie to fix neither
    
    
    double CL = 1.0;
    double tauKS = 0.08954, tauKL = 51.16; //in nano-seconds
    double MK = 497.614 /*in MeV/c^2*/,  dMK = 2.1890e-11 /*in MeV/c^2*/;

    double mom = 75000.0; //in MeV/c
    double c_light = 299.792; //in 10^6 m/s
    double tauK = 2.0/((1.0/tauKS)+(1.0/tauKL)), gamma = mom/MK;

    double lambdaL = gamma*c_light*tauKL, lambdaS = gamma*c_light*tauKS, lambda = gamma*c_light*tauK;

    ostringstream decayform, decayfit, testfit, testfit2;
    decayform<<CL<<"*exp(-x*"<<tauKS/tauKL<<") + CS*exp(-x*"<<tauKS/tauKS<<") + 2*exp(-x*"<<tauKS/tauK<<")*"<<DiluCon<<"*Cint*cos(Phi0-("<<dMK<<"*x/"<<tauKS<<"e-9))";                            
    //decayform<<CL<<"*exp(-x*"<<tauKS/tauKL<<")";
    //decayform<<"CS*exp(-x*"<<tauKS/tauKS<<")";
    //decayform<<CL<<"*exp(-x*"<<tauKS/tauKL<<") + CS*exp(-x*"<<tauKS/tauKS<<")";
    //decayform<<"2*exp(-x*"<<tauKS/tauK<<")*"<<DiluCon<<"*Cint*cos(Phi0-("<<dMK<<"*x/"<<tauKS<<"e-9))";
    //decayform<<CL<<"*exp(-x*"<<tauKS/tauKL<<") + 2*exp(-x*"<<tauKS/tauK<<")*"<<DiluCon<<"*Cint*cos(Phi0-("<<dMK<<"*x/"<<tauKS<<"e-9))";
    //decayform<<"CS*exp(-x*"<<tauKS/tauKS<<") + 2*exp(-x*"<<tauKS/tauK<<")*"<<DiluCon<<"*Cint*cos(Phi0-("<<dMK<<"*x/"<<tauKS<<"e-9))";

    
    //////////////////////////////////////////CLendd Evaluation////////////////////////////////////////////////////////
    ostringstream cldecayform;
    cldecayform<<CL<<"*exp(-x*"<<tauKS/tauKL<<")";
    TF1 *cldecayf = new TF1("cldecayf",cldecayform.str().c_str(),startt,endd);
    double CLendd = cldecayf->EvalPar(&endd);
    /////////////////////////////////////////////////////////////////////////////////////////////////////////////////////
    
    
    
    int i=0, j=0;

    RooRealVar x("x", "x", startt, endd);
    RooRealVar CS("CS", "CS", CSv, 0.00, 1.0);
    RooRealVar Cint("Cint", "Cint", Cintv, 0.00, 1.0);
    RooRealVar Phi0("Phi0", "Phi0", gRandom->Gaus(Phi0v,0.1), -1.58, 1.58);
    //Phi0.setConstant(true);
    RooGenericPdf decaypdf("decaypdf", "decaypdf", decayform.str().c_str(), RooArgSet(x, CS, Cint, Phi0));
    
    RooGaussian Phicon("Phicon","Phicon", Phi0, RooConst(Phi0v), RooConst(0.1/*From Precision Measurement Paper*/)) ;

    decayfit<<CL*CLfrac<<"*exp(-x*"<<tauKS/tauKL<<") + CS*exp(-x*"<<tauKS/tauKS<<") + 2*exp(-x*"<<tauKS/tauK<<")*"<<DiluCon<<"*Cint*cos(Phi0-("<<dMK<<"*x/"<<tauKS<<"e-9))";                                                       
    RooGenericPdf woaht("woaht", "woaht", decayfit.str().c_str(), RooArgSet(x, CS, Cint, Phi0));

    
    RooDataSet *data = decaypdf.generate(x, events);
    TH1 *roohist = data->createHistogram("roohist", x, Binning(bins,startt,endd));
    for(j=1;j<=bins;j++)
    {
        roohist->SetBinContent(j, (roohist->GetBinContent(j)-((CL-CLfrac)*roohist->GetBinContent(bins)*(CL/CLendd)*exp(-(j*(endd/bins))*tauKS/tauKL))));
        if(roohist->GetBinContent(j) < 0.0)
        {
            roohist->SetBinContent(j, 0.0);
        }
    }
    RooDataHist rdh("rdh", "rdh", x, roohist);

    
    TGraph csgr;
    TGraph cintgr;
    TGraph onlyint;
  
    for(i=0;i<(2*loops);i++)
    {
        sorint = (i%2);
        if(i==0)
        {
            CS.setVal(0.21);
            Cint.setVal(0.06);
        }
        if(sorint==1)
        {
            CS.setConstant(true);
            Cint.setConstant(false);
            //std::unique_ptr<RooFitResult> fitResult{decaypdf.fitTo(rdh, ExternalConstraints(Phicon), Save(), PrintLevel(-1))};
            std::unique_ptr<RooFitResult> fitResult{woaht.fitTo(rdh, ExternalConstraints(Phicon), Save(), PrintLevel(-1))};
            //std::unique_ptr<RooFitResult> fitResult{woaht.fitTo(*data, ExternalConstraints(Phicon), Save(), PrintLevel(-1))};
            cintgr.SetPoint((i/2), (i/2), (Cint.getValV()-Cintv)/Cintv);
            onlyint.SetPoint((i/2), (i/2), Cint.getValV());
        }
        if(sorint==0)
        {
            CS.setConstant(false);
            Cint.setConstant(true);
            //std::unique_ptr<RooFitResult> fitResult{decaypdf.fitTo(rdh, ExternalConstraints(Phicon), Save(), PrintLevel(-1))};
            std::unique_ptr<RooFitResult> fitResult{woaht.fitTo(rdh, ExternalConstraints(Phicon), Save(), PrintLevel(-1))};
            //std::unique_ptr<RooFitResult> fitResult{woaht.fitTo(*data, ExternalConstraints(Phicon), Save(), PrintLevel(-1))};
            csgr.SetPoint((i/2), (i/2), (CS.getValV()-CSv)/CSv);
        }
    }  
    
    
    TCanvas c("c","Canvase",1920,1080);
    
    cout<<"\n\n";
    csgr.SetTitle("Difference Between Fitted CS and True CS vs Recursion;Loop Number;(Fit(CS)-0.43)/0.43");
    csgr.Draw();
    gPad->Print("Recursive_CS.png");
    c.Clear();
    cintgr.SetTitle("Difference Between Fitted Cint and True Cint vs Recursion;Loop Number;(Fit(Cint)-0.12)/0.12");
    cintgr.Draw();
    gPad->Print("Recursive_Cint.png");
    c.Clear();
    onlyint.SetTitle("Fitted Cint vs Recursion;Loop Number;Fit(Cint)");
    onlyint.Draw();
    gPad->Print("Recursive_Cint_ActualVal.png");
    c.Clear();
    
    
    
    
    
    
    
    testfit<<CL*CLfrac<<"*exp(-x*"<<tauKS/tauKL<<") + "<<CSv<<"*exp(-x*"<<tauKS/tauKS<<") + 2*exp(-x*"<<tauKS/tauK<<")*"<<DiluCon<<"*"<<Cintv<<"*cos("<<Phi0v<<"-("<<dMK<<"*x/"<<tauKS<<"e-9))";                            
    RooGenericPdf testw("testw", "testw", testfit.str().c_str(), RooArgSet(x));

    testfit2<<CL*CLfrac<<"*exp(-x*"<<tauKS/tauKL<<") + "<<CSv<<"*exp(-x*"<<tauKS/tauKS<<")";                            
    RooGenericPdf testw2("testw", "testw", testfit2.str().c_str(), RooArgSet(x));

    
    
    RooPlot *xframe = x.frame(Title("K0 Decay PDF"));
    rdh.plotOn(xframe);
    //data->plotOn(xframe);
    woaht.plotOn(xframe, LineColor(kBlue));
    testw.plotOn(xframe, LineColor(kRed), LineStyle(kDashed));
    testw2.plotOn(xframe, LineColor(kGreen));
    xframe->Draw();
    gPad->Print("ROOHIST.png");
    
    
    
    
    
    
    
    
    
    std::array<double,2> carray = {CS.getValV(),Cint.getValV()};
    return carray;
}
```

```c++
//V O I D    M A I N
int bins = 60, noh = 1, events = (300000), loops = 100, i;
double DiluCon = 0.5;//0.285714286;   //Dilucon = 2.0/7.0;  N(K0)/N(K0bar) = 1.8
double CS = 0.43;//0.43
double Cint = 0.12;//0.12;
double Phi0 = 0.201357921;//0.201357921;


TH1D rechistcs("rechistcs","Recursive Optimization: C_{S};C_{S};Count",40,-0.0,1.0);
TH1D rechistcint("rechistcint","Recursive Optimization: C_{int};C_{int};Count",40,-0.0,1.0);
    
double startt = 0, endd = 6;
double CLfrac = 1.0;
std::array<double,2> carray;
for(i=0;i<noh;i++)
{
    carray = roof_CiPh2(bins, startt, endd, events, DiluCon, CS, Cint, Phi0, CLfrac, loops);
    rechistcs.Fill(carray[0]);
    rechistcint.Fill(carray[1]);
}

cout<<"\n\n";
gStyle->SetOptStat(1111111);
rechistcs.Draw();
gPad->Print("RecursiveHisto_CS.png");
rechistcint.Draw();
gPad->Print("RecursiveHisto_Cint.png");
```
